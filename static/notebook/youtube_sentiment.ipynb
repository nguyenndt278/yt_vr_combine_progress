```python
import pandas as pd
```

```python
df=pd.read_csv("C:\workspace\Resources\df_youtube_comments.csv")
df.head()
```

```python
sentiment_list=[]
for i in df['compound']:
    if float(i)<0:
        aa='negative'
    if float(i)>0:
        aa='possitive'
    if float(i)==0:
        aa='neutral'
    sentiment_list.append(aa)
print(sentiment_list)
```

```python
df['sentiment']=sentiment_list
df
```

```python
df.to_csv("C:\workspace\Resources\\youtube_sentiment.csv", index="false")
```
