```python
import pandas as pd
import isodate
```

```python
df_video=pd.read_csv("C:\workspace\Resources\df_youtube_viral_video_all.csv")
df_channel=pd.read_csv("C:\workspace\Resources\df_youtube_channel_statistics.csv")
df_category=pd.read_csv("C:\workspace\Resources\df_youtube_category.csv")
df_code=pd.read_csv("C:\workspace\Resources\df_country_code.csv")
```

```python
df_all_1=df_video.merge(df_category,how='left', on='Category')
df_all_2=df_all_1.merge(df_code,how='left', on='Code')
df_all=df_all_2.merge(df_channel, how='left', on='Channel ID')
df_all
```

```python
duration=df_all['Duration']
duration_list_raw=list(duration)
duration_list_parse=[]
for item in duration_list_raw:
    dura=isodate.parse_duration(item)
    durat= dura.total_seconds()
    duration_list_parse.append(durat)
df_all['Duration in Seconds']=duration_list_parse
df_tableau=df_all[["Name","id",
              "Title", "Published",
              "Channel", "Established",
              "Category Name", "Category",
              "Duration in Seconds", "Like",
              "Dislikes","Comment Count",
              "Channel Views", "Subscriber Count",
              "Video Count","Views"]]
df_tableau
```

```python
df_tableau_c=df_tableau.dropna()
df_tableau_clean=df_tableau_c.rename(columns={'Name':'Country'})
df_tableau_clean
```

```python
df_tableau_dummy=pd.get_dummies(df_tableau_clean, columns=["Country"])
df_tableau_dummy
```

```python
df_tableau_clean.to_csv("C:\workspace\Resources\df_youtube_tableau.csv", index="false")
df_tableau_dummy.to_csv("C:\workspace\Resources\df_youtube_tableau_dummy.csv", index="false")
```
