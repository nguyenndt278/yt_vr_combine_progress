```python
import pandas as pd
import isodate
```

```python
df_video=pd.read_csv("C:\workspace\Resources\df_youtube_viral_video_all.csv")
df_channel=pd.read_csv("C:\workspace\Resources\df_youtube_channel_statistics.csv")
df_channel_with_origin=pd.read_csv("C:\workspace\Resources\df_youtube_channel_statistics_with_origin.csv")
df_category=pd.read_csv("C:\workspace\Resources\df_youtube_category.csv")
df_code=pd.read_csv("C:\workspace\Resources\df_country_code.csv")
df_code_2=pd.read_csv("C:\workspace\Resources\df_country_code_2.csv")
df_sent=pd.read_csv("C:\workspace\Resources\youtube_sentiment.csv")
```

```python
df_sent
```

```python
df_all_1=df_video.merge(df_category,how='left', on='Category')
df_all_2=df_all_1.merge(df_code,how='left', on='Code')
df_all=df_all_2.merge(df_channel_with_origin, how='left', on='Channel ID')
df_all
```

```python
df_all_3=df_video.merge(df_category,how='left', on='Category')
df_all_4=df_all_3.merge(df_channel_with_origin, how='left', on='Channel ID')
df_all_origin=df_all_4.merge(df_code_2,how='left', on='Origin')
df_all_origin_c=df_all_origin.dropna()
df_all_origin_c
```

```python
df_all_origin_clean=df_all_origin_c[["Name","Origin","id",
              "Title", "Published",
              "Channel", "Established",
              "Category Name", "Category",
              "Duration", "Like",
              "Dislikes","Comment Count",
              "Channel Views", "Subscriber Count",
              "Video Count","Views"]]
df_all_origin_clean
```

```python
duration_2=df_all_origin_clean['Duration']
duration_list_raw_2=list(duration_2)
duration_list_parse_2=[]
for item in duration_list_raw_2:
    dura=isodate.parse_duration(item)
    durat= dura.total_seconds()
    duration_list_parse_2.append(durat)
df_all_origin_clean['Duration in Seconds']=duration_list_parse_2
df_all_origin_clean
```

```python
df_all_origin_clean_w=df_all_origin_clean.drop_duplicates()
df_all_origin_clean_w_na=df_all_origin_clean_w[["Name","Origin","id",
              "Title", "Published",
              "Channel", "Established",
              "Category Name", "Category",
              "Duration in Seconds", "Like",
              "Dislikes","Comment Count",
              "Channel Views", "Subscriber Count",
              "Video Count","Views"]]
df_all_origin_clean_w_name=df_all_origin_clean_w_na.rename(columns={'Name':'Country'})
df_all_origin_clean_w_name
```

```python
df_sentiment=df_all_origin_clean_w_name.merge(df_sent,how='left', on='id')
df_sentiment_c=df_sentiment.dropna()
df_sentiment_cle=df_sentiment_c.drop_duplicates()
df_sentiment_clean=df_sentiment_cle[['sentiment',
                                    'compound','neg',
                                    'neu','pos',
                                    'Views','Category',
                                    'Duration in Seconds',
                                    'Subscriber Count',
                                    'Video Count','Like','Dislikes',
                                     'Comment Count','Channel Views',
                                    'id','Origin']]
df_sentiment_dummy=pd.get_dummies(df_sentiment_clean, columns=["Origin"])
df_sentiment_dummy
```

```python
duration=df_all['Duration']
duration_list_raw=list(duration)
duration_list_parse=[]
for item in duration_list_raw:
    dura=isodate.parse_duration(item)
    durat= dura.total_seconds()
    duration_list_parse.append(durat)
df_all['Duration in Seconds']=duration_list_parse
df_tableau=df_all[["Name","Origin","id",
              "Title", "Published",
              "Channel", "Established",
              "Category Name", "Category",
              "Duration in Seconds", "Like",
              "Dislikes","Comment Count",
              "Channel Views", "Subscriber Count",
              "Video Count","Views"]]
df_tableau
```

```python
df_tableau_c=df_tableau.dropna()
df_tableau_cle=df_tableau_c.rename(columns={'Name':'Country'})
df_tableau_clean=df_tableau_cle.drop("Origin",1)
df_tableau_clean
```

```python
df_tableau_dummy=pd.get_dummies(df_tableau_clean, columns=["Country"])
df_tableau_dummy
```
